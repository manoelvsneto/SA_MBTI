```python
# ==============================
# Descriptive statistics + Boxplots
# By: Group, MBTI, GENDER
# Input: data.csv
# ==============================

import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# 1) Load file
# ------------------------------
FILE_PATH = "data.csv"

if not os.path.exists(FILE_PATH):
    print(f"Error: File {FILE_PATH} not found!")
else:
    df = pd.read_csv(FILE_PATH)
    print("File loaded:", FILE_PATH)
    print("Rows:", len(df))
    print("Columns:", list(df.columns))

# ------------------------------
# 2) Configuration
# ------------------------------
# Variables for grouping
GROUP_VARS = ["Group", "MBTI", "GENDER", "Profile"]

# Metrics to analyze (using exact column names from CSV)
# REQ = Nota Requisitos
# DEV = Nota Desenvolvimento  
# PROVA = Prova Teórica
# NOTA FINAL = Nota Final
METRICS = ["REQ", "DEV", "PROVA", "NOTA FINAL"]

# Normalize text in categorical columns
for c in GROUP_VARS:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Ensure numeric metrics
for c in METRICS:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# ------------------------------
# 3) Functions
# ------------------------------
def descriptive_by_group(data: pd.DataFrame, group_col: str, metric_col: str) -> pd.DataFrame:
    g = data[[group_col, metric_col]].dropna()
    if g.empty:
        return pd.DataFrame()

    def q1(x): return x.quantile(0.25)
    def q3(x): return x.quantile(0.75)

    out = (
        g.groupby(group_col)[metric_col]
         .agg(
            n="count",
            mean="mean",
            std="std",
            median="median",
            min="min",
            q1=q1,
            q3=q3,
            max="max"
         )
         .reset_index()
    )

    # round
    for col in ["mean", "std", "median", "min", "q1", "q3", "max"]:
        out[col] = out[col].astype(float).round(2)

    out = out.sort_values(by="n", ascending=False).reset_index(drop=True)
    return out

def sanitize_label(s: str) -> str:
    # remove problematic characters and make "latex-friendly"
    s = s.lower()
    s = s.replace("%", "pct")
    s = re.sub(r"[^\w\s-]", "", s)  # remove punctuation
    s = s.replace(" ", "-")
    # remove accents in a simple way (covers common PT-BR)
    repl = {
        "á":"a","à":"a","ã":"a","â":"a",
        "é":"e","ê":"e",
        "í":"i",
        "ó":"o","ô":"o","õ":"o",
        "ú":"u",
        "ç":"c"
    }
    for k,v in repl.items():
        s = s.replace(k, v)
    s = re.sub(r"-+", "-", s).strip("-")
    return s

def df_to_latex_table(df_in: pd.DataFrame, caption: str, label: str) -> str:
    """
    Generates a ready-to-use LaTeX block. Without 'booktabs' for compatibility.
    If you use \\usepackage{booktabs}, you can manually replace \\hline with \\toprule etc.
    """
    if df_in is None or df_in.empty:
        return f"% {caption}\n% (no data)\n"

    latex_tabular = df_in.to_latex(
        index=False,
        escape=True
    )

    latex = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        "\\small\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        f"{latex_tabular}\n"
        "\\end{table}\n"
    )
    return latex

def make_boxplot(data: pd.DataFrame, group_col: str, metric_col: str):
    d = data[[group_col, metric_col]].dropna()
    if d.empty:
        print(f"[WARNING] No data for boxplot: {metric_col} by {group_col}")
        return

    plt.figure(figsize=(12, 5))
    d.boxplot(column=metric_col, by=group_col, rot=45, grid=False)
    plt.title(f"{metric_col} by {group_col}")
    plt.suptitle("")
    plt.xlabel(group_col)
    plt.ylabel(metric_col)
    plt.tight_layout()
    plt.show()

# ------------------------------
# 4) Tables (screen + LaTeX)
# ------------------------------
metric_names = {
    "REQ": "Nota Requisitos",
    "DEV": "Nota Desenvolvimento",
    "PROVA": "Prova Teórica",
    "NOTA FINAL": "Nota Final"
}

for group_col in GROUP_VARS:
    if group_col not in df.columns:
        print(f"[WARNING] Categorical column not found and will be ignored: {group_col}")
        continue

    for metric in METRICS:
        if metric not in df.columns:
            print(f"[WARNING] Metric not found and will be ignored: {metric}")
            continue

        summary = descriptive_by_group(df, group_col, metric)

        print("\n" + "="*90)
        print(f"Descriptive statistics | {group_col} | {metric} ({metric_names.get(metric, metric)})")
        display(summary)

        metric_display = metric_names.get(metric, metric)
        latex_caption = f"Descriptive statistics of {metric_display} by {group_col}."
        latex_label = f"tab:desc-{sanitize_label(group_col)}-{sanitize_label(metric)}"
        latex_block = df_to_latex_table(summary, latex_caption, latex_label)

        print("\n--- LaTeX (copy and paste into paper) ---\n")
        print(latex_block)

# ------------------------------
# 5) Boxplots
# ------------------------------
print("\n" + "="*90)
print("Generating boxplots...")

for group_col in GROUP_VARS:
    if group_col not in df.columns:
        continue
    for metric in METRICS:
        if metric not in df.columns:
            continue
        make_boxplot(df, group_col, metric)

print("\nCompleted.")
```
