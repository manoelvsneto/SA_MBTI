```python
# ==============================
# GOOGLE COLAB - SCRIPT COMPLETO (CORRIGIDO)
# Estatística descritiva + Boxplots
# Por: Grupo, MBTI, GENDER
# Entrada: dados.xlsx
# ==============================

import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# 1) Carregar arquivo
# ------------------------------
FILE_PATH = "/content/dados.xlsx"

if not os.path.exists(FILE_PATH):
    from google.colab import files
    print("Faça upload do arquivo dados.xlsx ...")
    uploaded = files.upload()
    FILE_PATH = list(uploaded.keys())[0]

df = pd.read_excel(FILE_PATH)

print("Arquivo carregado:", FILE_PATH)
print("Linhas:", len(df))
print("Colunas:", list(df.columns))

# ------------------------------
# 2) Configurações
# ------------------------------
GROUP_VARS = ["Grupo", "MBTI", "GENDER"]

METRICS = [
    "Média Requisitos de Arquitetura",
    "Desenvolvimento Sistema",
    "PROVA OBJETIVA E AUTORIA",
    "NOTA_FINAL"  # vamos criar a partir da coluna original abaixo
]

NOTA_FINAL_COL = "NOTA FINAL GERAL ( TDE 40% + PROVA 60%)"

# Normalizar texto em colunas categóricas
for c in GROUP_VARS:
    if c in df.columns:
        df[c] = df[c].astype(str).str.strip()

# Criar coluna auxiliar NOTA_FINAL
if NOTA_FINAL_COL not in df.columns:
    raise ValueError(f"Coluna não encontrada: '{NOTA_FINAL_COL}'")

df["NOTA_FINAL"] = pd.to_numeric(df[NOTA_FINAL_COL], errors="coerce")

# Garantir métricas numéricas
numeric_cols = [
    "Média Requisitos de Arquitetura",
    "Desenvolvimento Sistema",
    "PROVA OBJETIVA E AUTORIA",
    "NOTA_FINAL",
]
for c in numeric_cols:
    if c not in df.columns:
        raise ValueError(f"Coluna não encontrada: '{c}'")
    df[c] = pd.to_numeric(df[c], errors="coerce")

# ------------------------------
# 3) Funções
# ------------------------------
def descriptive_by_group(data: pd.DataFrame, group_col: str, metric_col: str) -> pd.DataFrame:
    g = data[[group_col, metric_col]].dropna()
    if g.empty:
        return pd.DataFrame()

    def q1(x): return x.quantile(0.25)
    def q3(x): return x.quantile(0.75)

    out = (
        g.groupby(group_col)[metric_col]
         .agg(
            n="count",
            mean="mean",
            std="std",
            median="median",
            min="min",
            q1=q1,
            q3=q3,
            max="max"
         )
         .reset_index()
    )

    # arredondar
    for col in ["mean", "std", "median", "min", "q1", "q3", "max"]:
        out[col] = out[col].astype(float).round(2)

    out = out.sort_values(by="n", ascending=False).reset_index(drop=True)
    return out

def sanitize_label(s: str) -> str:
    # remove caracteres problemáticos e deixa "latex-friendly"
    s = s.lower()
    s = s.replace("%", "pct")
    s = re.sub(r"[^\w\s-]", "", s)  # remove pontuação
    s = s.replace(" ", "-")
    # remove acentos de forma simples (cobre PT-BR comum)
    repl = {
        "á":"a","à":"a","ã":"a","â":"a",
        "é":"e","ê":"e",
        "í":"i",
        "ó":"o","ô":"o","õ":"o",
        "ú":"u",
        "ç":"c"
    }
    for k,v in repl.items():
        s = s.replace(k, v)
    s = re.sub(r"-+", "-", s).strip("-")
    return s

def df_to_latex_table(df_in: pd.DataFrame, caption: str, label: str) -> str:
    """
    Gera um bloco LaTeX pronto. Sem 'booktabs' para compatibilidade.
    Se você usa \\usepackage{booktabs}, pode trocar manualmente \\hline por \\toprule etc.
    """
    if df_in is None or df_in.empty:
        return f"% {caption}\n% (sem dados)\n"

    latex_tabular = df_in.to_latex(
        index=False,
        escape=True
    )

    latex = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        "\\small\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        f"{latex_tabular}\n"
        "\\end{table}\n"
    )
    return latex

def make_boxplot(data: pd.DataFrame, group_col: str, metric_col: str):
    d = data[[group_col, metric_col]].dropna()
    if d.empty:
        print(f"[AVISO] Sem dados para boxplot: {metric_col} by {group_col}")
        return

    plt.figure(figsize=(12, 5))
    d.boxplot(column=metric_col, by=group_col, rot=45, grid=False)
    plt.title(f"{metric_col} by {group_col}")
    plt.suptitle("")
    plt.xlabel(group_col)
    plt.ylabel(metric_col)
    plt.tight_layout()
    plt.show()

# ------------------------------
# 4) Tabelas (tela + LaTeX)
# ------------------------------
for group_col in GROUP_VARS:
    if group_col not in df.columns:
        print(f"[AVISO] Coluna categórica não encontrada e será ignorada: {group_col}")
        continue

    for metric in METRICS:
        if metric not in df.columns:
            print(f"[AVISO] Métrica não encontrada e será ignorada: {metric}")
            continue

        summary = descriptive_by_group(df, group_col, metric)

        print("\n" + "="*90)
        print(f"Estatística descritiva | {group_col} | {metric}")
        display(summary)

        latex_caption = f"Estatística descritiva de {metric} por {group_col}."
        latex_label = f"tab:desc-{sanitize_label(group_col)}-{sanitize_label(metric)}"
        latex_block = df_to_latex_table(summary, latex_caption, latex_label)

        print("\n--- LaTeX (copiar e colar no artigo) ---\n")
        print(latex_block)

# ------------------------------
# 5) Boxplots
# ------------------------------
print("\n" + "="*90)
print("Gerando boxplots...")

for group_col in GROUP_VARS:
    if group_col not in df.columns:
        continue
    for metric in METRICS:
        if metric not in df.columns:
            continue
        make_boxplot(df, group_col, metric)

print("\nConcluído.")
```
